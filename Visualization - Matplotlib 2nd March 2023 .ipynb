Matplotlib is easy to use and an amazing visualizing library in Python.

It is built on NumPy arrays and designed to work with the broader SciPy stack and consists of several plots like line, bar, scatter, histogram, etc. 

```python
import matplotlib.pyplot as plt   # pythonic plot
```

```python
import numpy as np
import pandas as pd
```

```python
 x = np.linspace(1,100,200)
```

```python
x
```

```python
y = np.sin(x)
y
```

Plot Representation

```python
plt.plot(x,y)   # sin wave
```

```python
# labelling
plt.plot(x,y)
plt.xlabel('This is my xlabel')
plt.ylabel("this is y data")
plt.title('Sine graph')
plt.show()
```

```python
x = np.random.rand(50)

x
```

```python
y=np.random.rand(50)
y
```

```python
plt.scatter(x,y)
```

```python
# for different colours
colours = np.random.rand(50)
sizes = np.random.rand(50)
plt.scatter(x,y, c = colours , s = sizes)
```

```python
# for different colours  with size is increased
colours = np.random.rand(50)   # use colour code 
sizes = 1000 * np.random.rand(50)
plt.scatter(x,y, c = colours , s = sizes , alpha = 0.7)   # alpha - intensity of colour
```

```python
# for different colours  with size is increased
colours = np.random.rand(50)
sizes = 1000 * np.random.rand(50)
plt.scatter(x,y, c = colours , s = sizes , alpha = 0.1)
```

```python
# 
x = ['a','b','c','d','e']
```

```python
y = np.random.rand(5)
y
```

```python
# categorical vs numerical 
plt.bar(x,y)
plt.xlabel('categorical')
plt.ylabel('numerical')
plt.title("categorical vs numerical data by bar graph")
plt.show()
```

```python
# HORIZONTAL BAR Graph ------> barh
plt.barh(x,y)
plt.xlabel('numerical')
plt.ylabel('categorical')
plt.title("categorical vs numerical data by bar graph")
plt.show()
```

```python
plt.plot(x,y ,'--')
plt.xlabel('This is my xlabel')
plt.ylabel("this is y data")
plt.title('Sine graph')
plt.show()
```

```python
x= np.random.randint(-1,1,10)
y = np.random.randint(-1,1,10)
plt.figure(figsize=(6,2))
plt.plot(x,y,'--b',alpha = 0.9 , c ='b' )
plt.xlabel("this is x data")
plt.ylabel('this is y data')
plt.title('Title')
plt.grid()# grid is appearing at back
plt.show()
```

```python
# Histograph 
data = [1,3,4,5,3,5,3,6,4,8,9,6,8,5,9,4,7,4,1,2,3,4,5,6,9,4.5]

plt.xlabel('data distribution')
plt.ylabel('frequency of data')
plt.hist(data)
plt.show()
```

```python
s = [2,9,8,7,6,5]
p = [4,6,5,4,4,6]
colour = ["#77a940","#ffffff","#b57ab3","#4e6821","#685521","#98d05b"]

plt.scatter(s,p , c =colour,alpha = 0.9)
plt.show()
```

```python
# 3 D data
x = np.random.rand(100)
y = np.random.rand(100)
z = np.random.rand(100)
fig = plt.figure()
ax = fig.add_subplot(projection = '3d')
plt.scatter(x,y,z)
plt.show()
```
