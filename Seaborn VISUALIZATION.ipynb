#                      SEABORN 

Seaborn is a Python data visualization library based on matplotlib. It provides a high-level interface for drawing attractive and informative statistical graphics.

Seaborn is an amazing visualization library for statistical graphics plotting in Python. It provides beautiful default styles and color palettes to make statistical plots more attractive. It is built on top matplotlib library and is also closely integrated with the data structures from pandas.

```python
import seaborn as sns
```

```python
iris = sns.load_dataset('iris')
```

```python
iris['species'].unique()
```

```python
iris['sepal_width'].mean()
```

```python
# Relation b/w sepal_length & sepal_width
sns.scatterplot(x = iris.sepal_length , y = iris.sepal_width )
```

 Scatter plots are the graphs that present the relationship between two variables in a data-set. It represents data points on a two-dimensional plane or on a Cartesian system.

```python
# Relation b/w petal_length & petal_width
sns.scatterplot(x = iris.petal_length , y = iris.petal_width )
```

# Histogram 
A histogram is the graphical representation of data where data is grouped into continuous number ranges and each range corresponds to a vertical bar.



The horizontal axis displays the number range.


The vertical axis (frequency) represents the amount of data that is present in each range.


it is equivalent to displot in seaborn

```python
sns.distplot(iris["sepal_length"])
```

```python
sns.distplot(iris["petal_width"])
```

```python
tips = sns.load_dataset("tips")
```

Relational plots are used for visualizing the statistical relationship between the data points.




```python
# Relation plot  
sns.relplot(x = tips.total_bill , y = tips.tip ,data = tips ,hue = 'smoker')
```

```python
# categorical plot  - catplot
sns.catplot(x = 'day' , y = 'total_bill' , data  = tips)

# on saturday total_bill is high so management need to done prior
```

```python
sns.catplot(x = 'smoker',y = 'tip',data = tips)
```

```python
# joint plot 
sns.jointplot(x = tips.total_bill , y = tips.tip)
```

```python
# Pair plot 
sns.pairplot(tips)
```

```python
sns.pairplot(iris)
```

Violinplot
It is similar to the boxplot except that it provides a higher, more advanced visualization and uses the kernel density estimation to give a better description about the data distribution. 

```python
sns.violinplot(x ='day', y ='total_bill', data = tips, hue ='smoker', split = True)
```

```python
# Bivariate Histogram
sns.histplot(x = 'total_bill',y = 'tip',data = tips)
```

# Matrix Plot

Heatmap
Heatmap is defined as a graphical representation of data using colors to visualize the value of the matrix.

Syntax: seaborn.heatmap(data, *, vmin=None, vmax=None, cmap=None, center=None, annot_kws=None, linewidths=0, linecolor=â€™whiteâ€™, cbar=True, **kwargs)




clustermap



```python
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
data = np.random.randint(low = 1,
                         high = 100,
                         size = (10, 10))
print("The data to be plotted:\n")
print(data)
  
# plotting the heatmap
hm = sns.heatmap(data = data)
  
# displaying the plotted heatmap
plt.show()
```

```python
#The clustermap() function of seaborn plots a hierarchically-clustered heat map of the given matrix dataset. 
#It returns a clustered grid index. 
sns.clustermap(data, figsize=(7, 7))
plt.show()
```
