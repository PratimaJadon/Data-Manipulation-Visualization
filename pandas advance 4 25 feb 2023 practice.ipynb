```python
# apply function 
df['Fare_inr'] = df['Fare'].apply(lambda x : x*80)
df
```

```python
def euro_inr(x):
    return x*80 
df['Fare_inr'] = df['Fare'].apply(euro_inr)
df
```

```python
# categorise fare in 3 categories
def cat_fare(x):
    if x<10:
        return 'LOW'
    elif x>=10 and x<20 :
        return 'mid'
    else :
        return 'HIGH'
df['Fare_category'] = df['Fare'].apply(cat_fare)
df
```
